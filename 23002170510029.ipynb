```python
import math as mt
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

class UnitConversion:
    def __init__(self):
        while True:
            try:
                print("\nUnit Conversions:")
                print("1. Meter to Centimeter")
                print("2. Foot to Inches")
                print("3. Kilogram to Gram")
                print("4. Minute to Second")
                print("5. Celcius to Fahrenheit")
                print("6. Miles per Hour to Km per Hour")
                print("7. Litre to Ml")
                print("8. Square Meter To Square Feet")
                print("9. Centimeter To Meter")
                print("10. Inches to Foot")
                print("11. Gram to Kilogram")
                print("12. Second to Minute")
                print("13. Fahrenheit to Celcius")
                print("14. Km per Hour to Miles per Hour")
                print("15. Ml to Litre")
                print("16. Square Feet to Square Meter")
                print("0. Exit")

                choice = int(input("Enter your choice: "))
                if choice == 1:
                    print(self.mtcm(int(input("Enter Meter Value"))))
                elif choice == 2:
                    print(self.ftin(int(input("Enter Foot Value"))))
                elif choice == 3:
                    print(self.kgtg(int(input("Enter KG Value"))))
                elif choice == 4:
                    print(self.mts(int(input("Enter Minute Value"))))
                elif choice == 5:
                    print(self.ctf(int(input("Enter Celcius Value"))))
                elif choice == 6:
                    print(self.mphtkmph(int(input("Enter Mile per Hour Value"))))
                elif choice == 7:
                    print(self.ltml(int(input("Enter Litre Value"))))
                elif choice == 8:
                    print(self.smtsf(int(input("Enter Square Meter Value"))))
                elif choice == 9:
                    print(self.cmtm(int(input("Enter CentiMeter Value"))))
                elif choice == 10:
                    print(self.intf(int(input("Enter Inches Value"))))
                elif choice == 11:
                    print(self.gtkg(int(input("Enter Gram Value"))))
                elif choice == 12:
                    print(self.stm(int(input("Enter Seconds Value"))))
                elif choice == 13:
                    print(self.ftc(int(input("Enter Fahrenheit Value"))))
                elif choice == 14:
                    print(self.kmphtmph(int(input("Enter Km per hour Value"))))
                elif choice == 15:
                    print(self.mltl(int(input("Enter Millilitre Value"))))
                elif choice == 16:
                    print(self.sftsm(int(input("Enter Square Feet Value"))))
                elif choice == 0:
                    break
                else:
                    print("Invalid choice! Please select from 0 to 12.")
            except ValueError:
                print("Invalid input! Please enter a valid number.")

    def mtcm(self,m):
        return m*100
        
    def cmtm(self,cm):
        return cm/100
        
    def ftin(self,f):
        return f*12
        
    def intf(self,inc):
        return inc/12
        
    def kgtg(self,kg):
        return kg*1000

    def gtkg(self,g):
        return g/1000

    def mts(self,m):
        return m*60

    def stm(self,s):
        return s/60

    def ctf(self,c):
        return (c*9/5)+32

    def ftc(self,f):
        return (f-32)*5/9

    def mphtkmph(self,mph):
        return mph*1.60934

    def kmphtmph(self,kmph):
        return kmph*0.621371

    def ltml(self,l):
        return l*1000

    def mltl(self,ml):
        return ml/1000

    def smtsf(self,sm):
        return sm*10.764

    def sftsm(self,sf):
        return sm/10.764

class BasicCalculation:
    def __init__(self):
        while True:
            try:
                print("\nBasic Calculations:")
                print("1. Addition")
                print("2. Subtraction")
                print("3. Multiplication")
                print("4. Division")
                print("5. Logarithm")
                print("6. Factorial")
                print("7. Square Root")
                print("8. Power")
                print("9. Percentage")
                print("10. Exponential")
                print("11. Radian")
                print("12. Degree")
                print("0. Exit")

                choice = int(input("Enter your choice: "))
                if choice == 1:
                    print(self.addition())
                elif choice == 2:
                    print(self.subtraction())
                elif choice == 3:
                    print(self.multiplication())
                elif choice == 4:
                    print(self.division())
                elif choice == 5:
                    print(self.logarithm())
                elif choice == 6:
                    print(self.factorial())
                elif choice == 7:
                    print(self.square_root())
                elif choice == 8:
                    print(self.power())
                elif choice == 9:
                    print(self.percentage())
                elif choice == 10:
                    print(self.exponential())
                elif choice == 11:
                    print(self.get_radians())
                elif choice == 12:
                    print(self.get_degrees())
                elif choice == 0:
                    break
                else:
                    print("Invalid choice! Please select from 0 to 12.")
            except ValueError:
                print("Invalid input! Please enter a valid number.")

    def addition(self):
        try:
            numbers = list(map(float, input("Enter numbers separated by space: ").split()))
            return sum(numbers)
        except ValueError:
            return "Invalid input! Please enter only numbers."

    def subtraction(self):
        try:
            a = float(input("Enter first number: "))
            b = float(input("Enter second number: "))
            return a - b
        except ValueError:
            return "Invalid input! Please enter valid numbers."

    def multiplication(self):
        try:
            a = float(input("Enter first number: "))
            b = float(input("Enter second number: "))
            return a * b
        except ValueError:
            return "Invalid input! Please enter valid numbers."

    def division(self):
        try:
            a = float(input("Enter numerator: "))
            b = float(input("Enter denominator: "))
            if b == 0:
                raise ZeroDivisionError("Cannot divide by zero!")
            return a / b
        except ValueError:
            return "Invalid input! Please enter valid numbers."
        except ZeroDivisionError as e:
            return e

    def logarithm(self):
        try:
            base = float(input("Enter the base: "))
            number = float(input("Enter the number: "))
            if base <= 0 or number <= 0:
                raise ValueError("Base and number must be greater than zero.")
            return mt.log(number, base)
        except ValueError as e:
            return e

    def factorial(self):
        try:
            num = int(input("Enter a non-negative integer: "))
            if num < 0:
                raise ValueError("Factorial is not defined for negative numbers.")
            return mt.factorial(num)
        except ValueError as e:
            return e

    def square_root(self):
        try:
            num = float(input("Enter a number: "))
            if num < 0:
                raise ValueError("Cannot compute square root of negative number.")
            return mt.sqrt(num)
        except ValueError as e:
            return e

    def power(self):
        try:
            num = float(input("Enter the number: "))
            exp = float(input("Enter the exponent: "))
            return num ** exp
        except ValueError:
            return "Invalid input! Please enter valid numbers."

    def percentage(self):
        try:
            num = float(input("Enter the number: "))
            perc = float(input("Enter the percentage: "))
            return (num * perc) / 100
        except ValueError:
            return "Invalid input! Please enter valid numbers."

    def exponential(self):
        try:
            exp = float(input("Enter the exponent: "))
            return mt.exp(exp)
        except ValueError:
            return "Invalid input! Please enter a valid number."

    def get_radians(self):
        try:
            degree = float(input("Enter degrees: "))
            return mt.radians(degree)
        except ValueError:
            return "Invalid input! Please enter a valid number."

    def get_degrees(self):
        try:
            radian = float(input("Enter radians: "))
            return mt.degrees(radian)
        except ValueError:
            return "Invalid input! Please enter a valid number."

class StatisticsCalculation:
    def __init__(self):
        while True:
            try:
                print("\nStatistical Calculations:")
                print("1. Mean")
                print("2. Median")
                print("3. Mode")
                print("4. Variance")
                print("5. Standard Deviation")
                print("0. Exit")

                self.choice = int(input("Enter your choice: "))
                if self.choice==1:
                    print(self.meanof(list(map(float,input("Enter numbers but leave space between").split()))))
                elif self.choice==2:
                    self.medianof(list(map(float,input("Enter numbers but leave space between").split())))
                elif self.choice==3:
                    self.modeof(list(map(float,input("Enter numbers but leave space between").split())))
                elif self.choice==4:
                    print(self.varianceof(list(map(float,input("Enter numbers but leave space between").split()))))
                elif self.choice==5:
                    self.standarddeviationof(list(map(float,input("Enter numbers but leave space between").split())))
                elif self.choice == 0:
                    break
                else:
                    print("Invalid choice! Please select from 0 to 5.")
            except ValueError:
                print("Invalid input! Please enter a valid number.")

    def meanof(self,l):
        x=sum(l)/(len(l))
        return x

    def medianof(self,l):
        l.sort()
        if len(l)%2==0:
            x=((l[int((len(l))/2)-1])+(l[int((len(l))/2)]))/2 
        else:
            x=l[int((len(l)+1)/2)-1]
        print("Median:",x)

    def modeof(self,l):
        d={}
        max=0
        for i in l:
            if i in d:
                d[i]+=1
            else:
                d[i]=1
        for i in d:
            if d[i]>max:
                max=d[i]
        for i in d:
            if d[i]==max:
                print("Mode:",i)

    def varianceof(self,l):
        x=self.meanof(l)
        sum=0
        for i in l:
            sum+=(i-x)**2
        return sum/len(l)
        
    def standarddeviationof(self,l):
        print("Standard deviation:",self.varianceof(l)**0.5)
        
class EquationSolver:
    def solve_equation(self):
        try:
            print("\nEquation Solver (use 'x' as the variable)")
            print("\n** Use this sign for giving power or raise to")
            print("\nThis calculator is for one variable only so give one variable in the equation else it will give problems")
            equation = input("Enter an equation: ")
            x = sp.symbols('x')
            expr = sp.sympify(equation)
            solutions = sp.solve(expr, x)
            return f"Solutions: {solutions}"
        except (sp.SympifyError, ValueError):
            return "Invalid equation! Please enter a valid mathematical expression."


class GraphPlotter:
    def plot_graph(self):
        try:
            print("\nGraph Plotter (use 'x' as the variable)")
            print("\n** Use this sign for giving power or raise to")
            print("\nThis calculator is for one variable only so give one variable in the equation else it will give problems")
            equation = input("Enter an equation: ")
            x = sp.symbols('x')
            expr = sp.sympify(equation)
            f = sp.lambdify(x, expr, 'numpy')

            x_vals = np.linspace(-10, 10, 400)
            y_vals = f(x_vals)

            plt.plot(x_vals, y_vals)
            plt.title(f"Graph of {equation}")
            plt.xlabel('x')
            plt.ylabel('f(x)')
            plt.grid(True)
            plt.show()
        except Exception as e:
            print(f"Error plotting graph: {e}")


# Main Menu
equation_solver = EquationSolver()
graph_plotter = GraphPlotter()

while True:
    try:
        print("\nMain Menu:")
        print("1. Basic Calculations")
        print("2. Solve Equations")
        print("3. Plot Graph")
        print("4. Statistics")
        print("5. Unit Conversions")
        print("0. Exit")
        
        choice = int(input("Enter your choice: "))

        if choice == 1:
            BasicCalculation()
        elif choice == 2:
            print(equation_solver.solve_equation())
        elif choice == 3:
            graph_plotter.plot_graph()
        elif choice==4:
            StatisticsCalculation()
        elif choice==5:
            UnitConversion()
        elif choice == 0:
            print("Exiting the program. Goodbye!")
            break
        else:
            print("Invalid choice! Please enter a number between 0 and 4.")
    except ValueError:
        print("Invalid input! Please enter a valid number.")
```
